```python
import pandas as pd

# 1. Load the data
df = pd.read_csv("experiment_like_paper.csv")

# 2. Combine Algorithm and Assignment_Type to create unique column headers
df['Algo_Name'] = df['Algorithm'] + " (Type " + df['Assignment_Type'].astype(str) + ")"

# 3. Pivot the DataFrame
# Rows: Message_Count
# Columns: Algo_Name
# Values: Success_Rate
pivot_df = df.pivot(index="Message_Count", columns="Algo_Name", values="Success_Rate")

# Optional: Multiply by 100 if you want the values represented as percentages (e.g., 85.0 instead of 0.85)
pivot_df = pivot_df * 100

# 4. Generate the LaTeX table
# Note: In newer versions of Pandas (>= 1.3.0), to_latex() on the dataframe is deprecated 
# in favor of df.style.to_latex(). If you are using an older version, use pivot_df.to_latex(...)
latex_table = pivot_df.style.format("{:.2f}\\%").to_latex(
    caption="Success Percentage by Algorithm and Message Count",
    label="tab:success_percentage",
    position="htbp"
)

# Print or save the LaTeX code
print(latex_table)

# If you want to write it directly to a .tex file:
# with open("success_table.tex", "w") as f:
#     f.write(latex_table)
```
